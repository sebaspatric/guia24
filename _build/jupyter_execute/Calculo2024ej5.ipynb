Sea $a \in \left( -\dfrac{\pi}{2}, \dfrac{\pi}{2} \right)$.
¿Cuál es el valor de la siguiente integral?

$$
\int_{a}^{-a} \frac{\tan(x)}{x^2 + 1} \, dx
$$

## Desarrollo
Se pide calcular Integral:

***

### Paso 1: Usar la propiedad de cambio de límites

Recordemos la propiedad fundamental de las integrales definidas:

$$
\int_{a}^{-a} f(x)\,dx = - \int_{-a}^{a} f(x)\,dx
$$

Entonces:

$$
\int_{a}^{-a} \frac{\tan(x)}{x^2 + 1} \, dx = - \int_{-a}^{a} \frac{\tan(x)}{x^2 + 1} \, dx
$$

***

### Paso 2: Analizar la simetría del integrando

Sea

$$
f(x) = \frac{\tan(x)}{x^2 + 1}
$$

Veamos si es **par**, **impar** o ninguna:

* $\tan(-x) = -\tan(x)$
* $(-x)^2 + 1 = x^2 + 1$

Entonces:

$$
f(-x) = \frac{\tan(-x)}{(-x)^2 + 1} = \frac{-\tan(x)}{x^2 + 1} = -f(x)
$$

Por lo tanto, **la función es impar**.

***

### Paso 3: Propiedad de la integral de una función impar en intervalo simétrico

Si $f(x)$ es impar, entonces:

$$
\int_{-a}^{a} f(x) \, dx = 0
$$

Entonces:

$$
\int_{a}^{-a} f(x) \, dx = - \int_{-a}^{a} f(x) \, dx = -0 = 0
$$

***

### Conclusión final

$$
\int_{a}^{-a} \frac{\tan(x)}{x^2 + 1} \, dx = 0
$$

***
### Contexto

La función que estamos estudiando es:

$$
f(x) = \frac{\tan(x)}{x^2 + 1}
$$

* $\tan(x)$ es función **impar**: $\tan(-x) = -\tan(x)$
* $x^2 + 1$ es función **par**: $(-x)^2 + 1 = x^2 + 1$

Entonces, el cociente de una función impar sobre una función par:

$$
f(-x) = \frac{\tan(-x)}{(-x)^2 + 1} = \frac{-\tan(x)}{x^2 + 1} = -f(x)
$$

Por lo tanto, **$f(x)$ es función impar**.

***
### Integral definida de una función impar sobre intervalo simétrico

Para una función impar $f$, se cumple:

$$
\int_{-a}^{a} f(x) \, dx = 0
$$

Esto sucede porque el área bajo la curva de $f(x)$ desde $-a$ hasta 0 es el **negativo** del área desde 0 hasta $a$, y se cancelan entre sí.

***
### Interpretación

* Cuando decimos que $\displaystyle \int_{-a}^{a} f(x) \, dx = 0$, nos referimos a que **el área neta es cero** — las partes positivas y negativas se compensan.
* Sin embargo, la gráfica de la función $f(x)$ puede tener valores positivos y negativos.
* La gráfica de la integral acumulada

$$
F(x) = \int_0^x f(t) \, dt
$$

muestra cómo se acumula el área desde 0 hasta $x$:

* Para $x > 0$, $F(x)$ muestra la suma acumulada del área bajo $f$ desde 0 hasta $x$.
* Para $x < 0$, $F(x)$ muestra la suma acumulada desde 0 hacia la izquierda (áreas negativas porque $f$ es impar).

Así que aunque la integral simétrica desde $-a$ hasta $a$ sea cero, la integral acumulada $F(x)$ **no es cero para valores individuales de $x$**; solo se anula cuando se evalúa el área completa simétrica de $-a$ a $a$.

***
### Resumen

* La integral $\int_{-a}^{a} f(x) \, dx = 0$ porque $f$ es impar.
* La integral acumulada $F(x) = \int_0^x f(t) \, dt$ varía según $x$, mostrando áreas parciales que pueden ser positivas o negativas.
* Solo cuando $x = a$ y calculas la integral de $-a$ a $a$ completa, el resultado es 0.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import ipywidgets as widgets
from ipywidgets import interact

# Definir la función (impar)
def f(x):
    return np.tan(x) / (x**2 + 1)

# Dominio para graficar evitando ±pi/2
x = np.linspace(-np.pi/2 + 0.05, np.pi/2 - 0.05, 1000)
y = f(x)

# Función interactiva
def plot_areas_desde_menos_a(a=1.52):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9,10))
    fig.subplots_adjust(hspace=0.3)

    # ----------- Gráfico 1: función y áreas -----------
    ax1.plot(x, y, color="blue", label=r"$f(x)=\frac{\tan(x)}{x^2+1}$")
    ax1.axhline(0, color="black", lw=0.8)
    ax1.axvline(0, color="black", lw=0.8)

    # Líneas en -pi/2 y pi/2
    ax1.axvline(-np.pi/2, color="red", linestyle="--", lw=1, label=r"$x=-\frac{\pi}{2}$")
    ax1.axvline(np.pi/2, color="red", linestyle="--", lw=1, label=r"$x=\frac{\pi}{2}$")

    # Área izquierda [-a,0] en verde y área derecha [0,a] en naranja
    ax1.fill_between(x, 0, y, where=(x>=-abs(a)) & (x<=0), color="green", alpha=0.4,
                     label=rf"$\int_{{-{abs(a):.2f}}}^0 f(x)\,dx$")
    ax1.fill_between(x, 0, y, where=(x>=0) & (x<=a), color="orange", alpha=0.4,
                     label=rf"$\int_0^{a:.2f} f(x)\,dx$")

    # Calcular áreas
    A_izq, _ = quad(f, -abs(a), 0)
    A_der, _ = quad(f, 0, a)
    suma_areas = A_izq + A_der

    ax1.set_title(rf"Áreas: Izq={A_izq:.4f}, Der={A_der:.4f}, Suma={suma_areas:.4f}")
    ax1.set_xlabel("x")
    ax1.set_ylabel("f(x)")
    ax1.legend()
    ax1.grid(True)

    # ----------- Gráfico 2: Integral acumulada desde -a -----------
    # Inicializar integral acumulada
    F_acum = np.zeros_like(x)
    x0 = -abs(a)
    idx0 = np.argmin(np.abs(x - x0))
    F_acum[idx0] = 0  # Asegurar que F(-a) = 0

    # Integral hacia la derecha desde -a
    for i in range(idx0+1, len(x)):
        F_acum[i] = F_acum[i-1] + (y[i] + y[i-1])/2 * (x[i] - x[i-1])

    # Integral hacia la izquierda desde -a
    for i in range(idx0-1, -1, -1):
        F_acum[i] = F_acum[i+1] - (y[i+1] + y[i])/2 * (x[i+1] - x[i])

    ax2.plot(x, F_acum, color="red", label=rf"$F(x)=\int_{{-{abs(a):.2f}}}^x f(t) dt$")
    ax2.axhline(0, color="black", lw=0.8)
    ax2.axvline(x0, color="green", lw=1, linestyle="--", label=rf"Inicio integral $x_0=-{abs(a):.2f}$")

    # Punto en x=a mostrando integral total de -a a a
    ax2.plot(a, suma_areas, "ro", markersize=8, label=rf"$F(a)={suma_areas:.4f}$")

    ax2.set_title("Integral acumulada desde -a")
    ax2.set_xlabel("x")
    ax2.set_ylabel("F(x)")
    ax2.legend()
    ax2.grid(True)

    plt.show()
plot_areas_desde_menos_a()
# Deslizador interactivo para a en [-pi/2+0.05, pi/2-0.05]
#interact(plot_areas_desde_menos_a, a=widgets.FloatSlider(value=1.52, min=0.05, max=np.pi/2-0.05, step=0.01))
```

### **Funciones definidas de $0$ a $\pi/2$**

1. **Corriente:**

$$
I(t) = \sin(t) \quad \text{(impar)}
$$

* Propiedad impar: $I(-t) = -I(t)$
* Área bajo la curva en $[0, \pi/2]$ positiva, en $[- \pi/2, 0]$ negativa.

2. **Voltaje:**

$$
V(t) = \cos(t) \quad \text{(par)}
$$

* Propiedad par: $V(-t) = V(t)$
* Área bajo la curva simétrica respecto a $t=0$.

3. **Potencia instantánea:**

$$
P(t) = I(t)\cdot V(t) = \sin(t)\cdot \cos(t) \quad \text{(impar)}
$$

* Producto de una función impar por una par → **impar**
* Área positiva para $t>0$, negativa para $t<0$.

4. **Energía acumulada aproximada:**

$$
E(t) = \int_0^t P(\tau)\, d\tau
$$

* Para $t \in [0, \pi/2]$ es positiva y crece con $t$.
* Si integráramos de $-\pi/2$ a $\pi/2$, se cancelaría por la simetría impar de $P(t)$.

---

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from scipy.integrate import quad

# Señales
def i(t):
    return np.sin(t)  # corriente impar

def v(t):
    return np.cos(t)  # voltaje par

# Dominio de tiempo
t = np.linspace(0, 2*np.pi, 1000)
dt = t[1] - t[0]

# Función interactiva
def plot_circuit_integral(a=np.pi/2):
    i_vals = i(t)
    v_vals = v(t)
    p_vals = i_vals * v_vals  # potencia instantánea

    # Energía acumulada usando regla del trapecio
    W = np.zeros_like(t)
    for k in range(1, len(t)):
        W[k] = W[k-1] + 0.5*(p_vals[k] + p_vals[k-1]) * dt

    # Valor de energía acumulada en t=a
    idx_a = np.argmin(np.abs(t - a))
    W_a = W[idx_a]

    fig, axes = plt.subplots(3, 1, figsize=(10,12))
    fig.subplots_adjust(hspace=0.4)

    # -------- Corriente --------
    axes[0].plot(t, i_vals, label=r"$i(t)=\sin(t)$", color="blue")
    axes[0].fill_between(t, 0, i_vals, where=(t<=a), color="orange", alpha=0.3)
    axes[0].axvline(a, color="red", linestyle="--", label=f"t = {a:.2f}")
    axes[0].set_title("Corriente i(t)")
    axes[0].set_xlabel("t")
    axes[0].set_ylabel("i(t)")
    axes[0].legend()
    axes[0].grid(True)

    # -------- Voltaje --------
    axes[1].plot(t, v_vals, label=r"$v(t)=\cos(t)$", color="green")
    axes[1].fill_between(t, 0, v_vals, where=(t<=a), color="purple", alpha=0.3)
    axes[1].axvline(a, color="red", linestyle="--", label=f"t = {a:.2f}")
    axes[1].set_title("Voltaje v(t)")
    axes[1].set_xlabel("t")
    axes[1].set_ylabel("v(t)")
    axes[1].legend()
    axes[1].grid(True)

    # -------- Potencia y energía acumulada --------
    axes[2].plot(t, p_vals, label=r"$p(t)=v(t)\cdot i(t)$", color="red")
    axes[2].fill_between(t, 0, p_vals, where=(t<=a), color="yellow", alpha=0.3)
    axes[2].plot(t, W, label=rf"Energía acumulada $W(t)=\int_0^t p(\tau)d\tau$", color="black", linestyle="--")
    axes[2].axvline(a, color="red", linestyle="--", label=f"t = {a:.2f}")

    # Marcar el valor de energía acumulada en t=a
    axes[2].plot(a, W_a, "ro", markersize=8)
    axes[2].text(a, W_a, f"W({a:.2f})={W_a:.4f}", color="red", fontsize=10, verticalalignment="bottom")

    axes[2].set_title("Potencia y energía acumulada")
    axes[2].set_xlabel("t")
    axes[2].set_ylabel("p(t) / W(t)")
    axes[2].legend()
    axes[2].grid(True)

    plt.show()

# Deslizador interactivo
interact(plot_circuit_integral, a=widgets.FloatSlider(value=np.pi/2, min=0, max=2*np.pi, step=0.01))
```

4. **Energía acumulada aproximada:**

¡Perfecto! 😄 Ahora vamos a definir la energía acumulada **sobre un intervalo simétrico $[-t, t]$**, para ver cómo se comporta la integral de potencia cuando la corriente es impar y el voltaje es par.

---

### **Definición de energía acumulada simétrica**

$$
E(t) = \int_{-t}^{t} P(\tau) \, d\tau = \int_{-t}^{t} I(\tau) V(\tau) \, d\tau
$$

Donde:

* $I(\tau) = \sin(\tau)$ (impar)
* $V(\tau) = \cos(\tau)$ (par)
* $P(\tau) = I(\tau) \cdot V(\tau)$ (impar)

**Observación:**

* Como $P(\tau)$ es **impar**, la integral sobre un intervalo simétrico alrededor de cero siempre será **cero**:

$$
E(t) = \int_{-t}^{t} P(\tau)\,d\tau = 0
$$

* Sin embargo, si quieres ver el **proceso de acumulación**, podemos graficar la integral **desde -t hasta cada punto positivo**, o mostrar áreas positivas y negativas separadas.


---

 **Qué se muestra:**

1. Área de potencia $P(t)$ desde $-a$ hasta $a$ sombreada en amarillo.
2. Líneas verticales indicando los límites $-a$ y $a$.
3. Valor de la energía acumulada $E(a)$, que será **0** para cualquier $a$ debido a la simetría impar.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Señales
def i(t):
    return np.sin(t)  # corriente impar

def v(t):
    return np.cos(t)  # voltaje par

# Dominio de tiempo desde -pi/2 hasta 2pi
t_start = -np.pi/2
t_end = 2*np.pi
t = np.linspace(t_start, t_end, 1000)
dt = t[1] - t[0]

# Función interactiva
def plot_circuit_integral(a=np.pi/2):
    i_vals = i(t)
    v_vals = v(t)
    p_vals = i_vals * v_vals  # potencia instantánea

    # Energía acumulada desde -pi/2
    W = np.zeros_like(t)
    for k in range(1, len(t)):
        W[k] = W[k-1] + 0.5*(p_vals[k] + p_vals[k-1]) * dt

    # Valor de energía acumulada en t=a
    idx_a = np.argmin(np.abs(t - a))
    W_a = W[idx_a]

    fig, axes = plt.subplots(3, 1, figsize=(10,12))
    fig.subplots_adjust(hspace=0.4)

    # -------- Corriente --------
    axes[0].plot(t, i_vals, label=r"$i(t)=\sin(t)$", color="blue")
    axes[0].fill_between(t, 0, i_vals, where=(t<=a), color="orange", alpha=0.3)
    axes[0].axvline(a, color="red", linestyle="--", label=f"t = {a:.2f}")
    axes[0].set_title("Corriente i(t)")
    axes[0].set_xlabel("t")
    axes[0].set_ylabel("i(t)")
    axes[0].legend()
    axes[0].grid(True)

    # -------- Voltaje --------
    axes[1].plot(t, v_vals, label=r"$v(t)=\cos(t)$", color="green")
    axes[1].fill_between(t, 0, v_vals, where=(t<=a), color="purple", alpha=0.3)
    axes[1].axvline(a, color="red", linestyle="--", label=f"t = {a:.2f}")
    axes[1].set_title("Voltaje v(t)")
    axes[1].set_xlabel("t")
    axes[1].set_ylabel("v(t)")
    axes[1].legend()
    axes[1].grid(True)

    # -------- Potencia y energía acumulada --------
    axes[2].plot(t, p_vals, label=r"$p(t)=v(t)\cdot i(t)$", color="red")
    axes[2].fill_between(t, 0, p_vals, where=(t<=a), color="yellow", alpha=0.3)
    axes[2].plot(t, W, label=rf"Energía acumulada $W(t)=\int_{{-\pi/2}}^t p(\tau)d\tau$", color="black", linestyle="--")

    # Línea vertical en el límite t=a
    axes[2].axvline(a, color="red", linestyle="--", label=f"Límite t = {a:.2f}")
    # Punto y valor de energía acumulada en el límite
    axes[2].plot(a, W_a, "ro", markersize=8)
    axes[2].text(a, W_a, f"W({a:.2f})={W_a:.4f}", color="red", fontsize=10, verticalalignment="bottom", horizontalalignment="left")

    axes[2].set_title("Potencia y energía acumulada")
    axes[2].set_xlabel("t")
    axes[2].set_ylabel("p(t) / W(t)")
    axes[2].legend()
    axes[2].grid(True)

    plt.show()

# Deslizador interactivo: valor inicial = pi/2
interact(plot_circuit_integral, a=widgets.FloatSlider(value=np.pi/2, min=t_start, max=t_end, step=0.01))
```

```python
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from ipywidgets import interact
from plotly.subplots import make_subplots   


# Señales
def i(t):
    return np.sin(t)  # corriente impar

def v(t):
    return np.cos(t)  # voltaje par

# Dominio de tiempo desde -pi/2 hasta 2pi
t_start = -np.pi/2
t_end = 2*np.pi
t = np.linspace(t_start, t_end, 1000)
dt = t[1] - t[0]

# Función interactiva
def plot_circuit_integral(a=np.pi/2):
    i_vals = i(t)
    v_vals = v(t)
    p_vals = i_vals * v_vals  # potencia instantánea

    # Energía acumulada desde -pi/2
    W = np.zeros_like(t)
    for k in range(1, len(t)):
        W[k] = W[k-1] + 0.5*(p_vals[k] + p_vals[k-1]) * dt

    # Valor de energía acumulada en t=a
    idx_a = np.argmin(np.abs(t - a))
    W_a = W[idx_a]

    # Crear subplots
    fig = make_subplots(rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.15,
                        subplot_titles=("Corriente i(t)", "Voltaje v(t)", "Potencia y energía acumulada"))

    # -------- Corriente --------
    fig.add_trace(go.Scatter(x=t, y=i_vals, name="i(t)=sin(t)", line=dict(color="blue")), row=1, col=1)
    fig.add_trace(go.Scatter(x=t[t<=a], y=i_vals[t<=a], fill='tozeroy', mode="lines",
                             line=dict(color="orange"), name="Área hasta t=a"), row=1, col=1)
    fig.add_vline(x=a, line=dict(color="red", dash="dash"), row=1, col=1)

    # -------- Voltaje --------
    fig.add_trace(go.Scatter(x=t, y=v_vals, name="v(t)=cos(t)", line=dict(color="green")), row=2, col=1)
    fig.add_trace(go.Scatter(x=t[t<=a], y=v_vals[t<=a], fill='tozeroy', mode="lines",
                             line=dict(color="purple"), name="Área hasta t=a"), row=2, col=1)
    fig.add_vline(x=a, line=dict(color="red", dash="dash"), row=2, col=1)

    # -------- Potencia y energía acumulada --------
    fig.add_trace(go.Scatter(x=t, y=p_vals, name="p(t)=v(t)·i(t)", line=dict(color="red")), row=3, col=1)
    fig.add_trace(go.Scatter(x=t[t<=a], y=p_vals[t<=a], fill='tozeroy', mode="lines",
                             line=dict(color="yellow"), name="Área hasta t=a"), row=3, col=1)
    fig.add_trace(go.Scatter(x=t, y=W, name="Energía acumulada W(t)", line=dict(color="black", dash="dash")), row=3, col=1)
    fig.add_vline(x=a, line=dict(color="red", dash="dash"), row=3, col=1)
    fig.add_trace(go.Scatter(x=[a], y=[W_a], mode="markers+text", text=[f"W({a:.2f})={W_a:.4f}"],
                             textposition="bottom right", marker=dict(color="red", size=8), name="W(a)"), row=3, col=1)

    # Layout
    fig.update_layout(height=900, width=900, title=f"Circuito: Corriente, Voltaje y Energía (t={a:.2f})",
                      showlegend=True, template="plotly_white")
    fig.show()

# Slider interactivo
interact(plot_circuit_integral,
         a=widgets.FloatSlider(value=np.pi/2, min=t_start, max=t_end, step=0.01, description="t"))
```
