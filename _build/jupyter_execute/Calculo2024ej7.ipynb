```python
### sucecion fibonacci
import numpy as np
import plotly.graph_objects as go

# Sucesi贸n Fibonacci
N_total_max = 50
fib = [1, 1]
for n in range(2, N_total_max):
    fib.append(fib[n-1] + fib[n-2])
fib = np.array(fib)
n_vals = np.arange(1, N_total_max+1)

# Crear figura con 3 traces
fig = go.Figure()

trace_an = go.Scatter(x=n_vals[:20], y=fib[:20], mode='lines+markers', name='a_n', line=dict(color='blue'))
trace_an1 = go.Scatter(x=n_vals[1:20], y=fib[:19], mode='lines+markers', name='a_{n-1}', line=dict(color='orange', dash='dash'))
trace_an2 = go.Scatter(x=n_vals[2:20], y=fib[:18], mode='lines+markers', name='a_{n-2}', line=dict(color='green', dash='dot'))

fig.add_traces([trace_an, trace_an1, trace_an2])

# Crear slider
steps = []
for N_max in range(2, N_total_max+1):
    x_an = n_vals[:N_max]
    y_an = fib[:N_max]

    x_an1 = n_vals[1:N_max] if N_max>1 else []
    y_an1 = fib[:N_max-1] if N_max>1 else []

    x_an2 = n_vals[2:N_max] if N_max>2 else []
    y_an2 = fib[:N_max-2] if N_max>2 else []

    step = dict(
        method='update',
        args=[{'x':[x_an, x_an1, x_an2],
               'y':[y_an, y_an1, y_an2]}],
        label=str(N_max)
    )
    steps.append(step)

sliders = [dict(active=18, pad={"t":50}, steps=steps, currentvalue={"prefix":"N_max = "})]

# Layout
fig.update_layout(
    autosize=True,
    width=None,
    height=None,
    sliders=sliders,
    title='Sucesi贸n Fibonacci con a_{n-1} y a_{n-2}',
    xaxis_title='n',
    yaxis_title='Valor',
    template='plotly_white'
)
fig.write_html(
    r'_build/html/7sucesion_fibonacci.html',
    include_plotlyjs='cdn',
    full_html=False
)
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from fractions import Fraction

# Funci贸n Fibonacci
def fibonacci(n):
    a = [1, 1]
    for i in range(2, n):
        a.append(a[i-1] + a[i-2])
    return np.array(a)

# N煤mero m谩ximo de t茅rminos
N_max = 20
a_full = fibonacci(N_max+1)
phi = (1 + np.sqrt(5)) / 2
n_vals_total = np.arange(1, N_max+1)

# Crear figura
fig = go.Figure()

# Inicialmente mostrar todos los t茅rminos
b_vals_frac_init = [Fraction(a_full[i+1], a_full[i]) for i in range(N_max)]
b_vals_float_init = [float(f) for f in b_vals_frac_init]
b_labels_init = [f"{f.numerator}/{f.denominator}  {float(f):.4f}" for f in b_vals_frac_init]

fig.add_trace(go.Scatter(
    x=n_vals_total,
    y=b_vals_float_init,
    mode='lines+markers',
    name='b_n',
    text=b_labels_init,
    hovertemplate="n=%{x}<br>b_n=%{text}<extra></extra>",
    line=dict(color='blue')
))

# L铆nea del n煤mero 谩ureo
fig.add_trace(go.Scatter(
    x=n_vals_total,
    y=[phi]*N_max,
    mode='lines',
    name='',
    line=dict(color='red', dash='dash')
))

# Crear pasos para el slider
steps = []
for N in range(2, N_max+1):
    b_frac = [Fraction(a_full[i+1], a_full[i]) for i in range(N)]
    b_float = [float(f) for f in b_frac]
    b_labels = [f"{f.numerator}/{f.denominator}  {float(f):.4f}" for f in b_frac]
    
    step = dict(
        method='update',
        args=[{
            'x': [n_vals_total[:N], n_vals_total[:N]],
            'y': [b_float, [phi]*N]
        },
        {
            'trace': [0, 1],
            'text': [b_labels, None]
        }],
        label=str(N)
    )
    steps.append(step)

# Agregar slider
sliders = [dict(
    active=N_max-1,
    currentvalue={"prefix":"N煤mero de t茅rminos: "},
    pad={"t":50},
    steps=steps
)]

fig.update_layout(
    autosize=True,
    width=None,
    height=None,
    sliders=sliders,
    title="Convergencia de b = a / a al n煤mero 谩ureo  = (1+5)/2  1.618",
    xaxis_title='n',
    yaxis_title='b ',
    template='plotly_white'
)
fig.write_html(r"_build/html/7numeroaureo.html",
               include_plotlyjs='cdn',
               full_html=False)

fig.show()
```
