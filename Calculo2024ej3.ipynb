## Pregunta3: continuidad

### **3. Encuentre los valores de $a$ y $b$ para que la funci√≥n $f(x)$ sea continua en $\mathbb{R}$:**

$$
f(x) =
\begin{cases}
\dfrac{x^2 - 4}{x - 2}, & \text{si } x < 2 \\
a x^2 - b x + 3, & \text{si } 2 \leq x \leq 3 \\
2x - a + b, & \text{si } x > 3
\end{cases}
$$



***
### **Desarrollo**

La funci√≥n $f(x)$ debe ser continua **en todo $\mathbb{R}$**.
Esto significa que no debe haber saltos en los puntos donde cambia la definici√≥n: $x = 2$ y $x = 3$.

***
### **Paso 1: Continuidad en $x = 2$**

Para que $f(x)$ sea continua en $x = 2$:

$$
\lim_{x \to 2^-} f(x) = f(2) = \lim_{x \to 2^+} f(x)
$$

#### L√≠mite por la izquierda ($x \to 2^-$)

$$
f(x) = \frac{x^2 - 4}{x - 2} = \frac{(x-2)(x+2)}{x-2} = x+2, \quad x\neq 2
$$

$$
\lim_{x \to 2^-} f(x) = 2 + 2 = 4
$$

#### Evaluar $f(2)$

$$
f(2) = a(2)^2 - b(2) + 3 = 4a - 2b + 3
$$

#### L√≠mite por la derecha ($x \to 2^+$)

$$
\lim_{x \to 2^+} f(x) = f(2) = 4a - 2b + 3
$$

#### Igualar l√≠mites

$$
4 = 4a - 2b + 3 \quad \Rightarrow \quad 4a - 2b = 1 \quad (‚ë†)
$$

***

### **Paso 2: Continuidad en $x = 3$**

Para continuidad en $x = 3$:

$$
\lim_{x \to 3^-} f(x) = f(3) = \lim_{x \to 3^+} f(x)
$$

#### L√≠mite por la izquierda ($x \to 3^-$)

$$
\lim_{x \to 3^-} f(x) = a(3)^2 - b(3) + 3 = 9a - 3b + 3
$$

#### L√≠mite por la derecha ($x \to 3^+$)

$$
\lim_{x \to 3^+} f(x) = 2(3) - a + b = 6 - a + b
$$

#### Igualar l√≠mites

$$
9a - 3b + 3 = 6 - a + b \quad \Rightarrow \quad 10a - 4b = 3 \quad (‚ë°)
$$

***

### **Paso 3: Resolver el sistema**

Sistema:

$$
\begin{cases}
4a - 2b = 1 \quad (‚ë†) \\
10a - 4b = 3 \quad (‚ë°)
\end{cases}
$$

Multiplicamos (‚ë†) por 2:

$$
8a - 4b = 2
$$

Restamos de (‚ë°):

$$
(10a - 4b) - (8a - 4b) = 3 - 2 \quad \Rightarrow \quad 2a = 1 \quad \Rightarrow a = \frac{1}{2}
$$

Sustituimos $a = \frac{1}{2}$ en (‚ë†):

$$
4 \cdot \frac{1}{2} - 2b = 1 \quad \Rightarrow \quad 2 - 2b = 1 \quad \Rightarrow \quad b = \frac{1}{2}
$$

***

### **Soluci√≥n final**

$$
\boxed{a = \frac{1}{2}, \quad b = \frac{1}{2}}
$$

***

#### **Respuesta correcta**

**c) $a = \dfrac{1}{2}, \; b = \dfrac{1}{2}$**

```python
#C√≥digo para graficar la funci√≥n por partes continua en ‚Ñù
import numpy as np
import plotly.graph_objects as go

# Funciones por tramos
def f_left(x):
    return (x**2 - 4) / (x - 2)

def f_middle(x, a, b):
    return a * x**2 - b * x + 3

def f_right(x, a, b):
    return 2 * x - a + b

# Valores iniciales
a_init, b_init = 0.5, 0.5

# Rango de cada tramo
x1 = np.linspace(1.5, 1.99, 200)
x2 = np.linspace(2, 3, 200)
x3 = np.linspace(3.01, 3.5, 200)

# Crear figura
fig = go.Figure()

# Trazas iniciales
fig.add_trace(go.Scatter(x=x1, y=f_left(x1), mode="lines", name="x < 2", line=dict(color="blue")))
fig.add_trace(go.Scatter(x=x2, y=f_middle(x2, a_init, b_init), mode="lines", name="2 ‚â§ x ‚â§ 3", line=dict(color="green")))
fig.add_trace(go.Scatter(x=x3, y=f_right(x3, a_init, b_init), mode="lines", name="x > 3", line=dict(color="red")))

# Puntos en x=2 y x=3
fig.add_trace(go.Scatter(x=[2], y=[f_middle(2, a_init, b_init)], mode="markers", name="f(2)", marker=dict(color="black", size=8)))
fig.add_trace(go.Scatter(x=[3], y=[f_middle(3, a_init, b_init)], mode="markers", name="f(3)", marker=dict(color="black", size=8)))

# Sliders para 'a' y 'b'
a_values = np.linspace(-2, 3, 21)
b_values = np.linspace(-2, 3, 21)

steps_a = []
for aval in a_values:
    steps_a.append(dict(
        method="update",
        label=str(round(aval,2)),
        args=[{"y": [
            f_left(x1),                        # x<2 no depende de a,b
            f_middle(x2, aval, b_init),        # tramo medio
            f_right(x3, aval, b_init),         # tramo derecho
            [f_middle(2, aval, b_init)],       # punto en x=2
            [f_middle(3, aval, b_init)]        # punto en x=3
        ]}]
    ))

steps_b = []
for bval in b_values:
    steps_b.append(dict(
        method="update",
        label=str(round(bval,2)),
        args=[{"y": [
            f_left(x1),
            f_middle(x2, a_init, bval),
            f_right(x3, a_init, bval),
            [f_middle(2, a_init, bval)],
            [f_middle(3, a_init, bval)]
        ]}]
    ))

sliders = [
    dict(active=int(len(a_values)/2), currentvalue={"prefix": "a = "},
         pad={"t": 50}, steps=steps_a, x=0, xanchor="left", y=-0.15, len=0.48),
    dict(active=int(len(b_values)/2), currentvalue={"prefix": "b = "},
         pad={"t": 50}, steps=steps_b, x=0.52, xanchor="left", y=-0.15, len=0.48)
]

# Layout
fig.update_layout(
    title="Funci√≥n por partes continua en ‚Ñù",
    xaxis_title="x",
    yaxis_title="f(x)",
    sliders=sliders,
    template="plotly_white",
    margin=dict(b=150)
)

fig.show()
```

### gr√°fica est√°tica

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de a y b encontrados
a = 0.5
b = 0.5

# Definimos las tres partes de la funci√≥n
def f_left(x):
    return (x**2 - 4) / (x - 2)

def f_middle(x):
    return a * x**2 - b * x + 3

def f_right(x):
    return 2 * x - a + b

# Crear rangos de x para cada tramo
x1 = np.linspace(1.5, 1.99, 200)
x2 = np.linspace(2, 3, 200)
x3 = np.linspace(3.01, 3.5, 200)

# Calcular y para cada tramo
y1 = f_left(x1)
y2 = f_middle(x2)
y3 = f_right(x3)

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar cada tramo
plt.plot(x1, y1, label='x < 2', color='blue')
plt.plot(x2, y2, label='2 ‚â§ x ‚â§ 3', color='green')
plt.plot(x3, y3, label='x > 3', color='red')

# Puntos exactos en x = 2 y x = 3
plt.plot(2, f_middle(2), 'ko', label='f(2)')
plt.plot(3, f_middle(3), 'ko', label='f(3)')

# L√≠neas verticales para guiar visualmente
plt.axvline(x=2, color='gray', linestyle='--', alpha=0.4)
plt.axvline(x=3, color='gray', linestyle='--', alpha=0.4)

# Est√©tica del gr√°fico
plt.title('Funci√≥n por partes continua en ‚Ñù')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```
